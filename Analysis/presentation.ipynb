<h1 style="text-align:center;">Introduction</h1>

Predictive analytics is a powerful approach that leverages **historical data, statistical algorithms, and machine learning** to forecast future outcomes. 

- By identifying patterns and trends within existing data, organisations can make informed, proactive decisions.
- Predictive analytics is widely applied across industries such as finance, healthcare, retail, telecommunications, and manufacturing

## 1. Define the Goal
The first step is to clearly define the business problem you want to solve. This is crucial because it guides the entire process. 

- Are you trying to predict which customers are likely to stop using your service? Or do you want to forecast product demand for the next quarter?
- A specific question leads to a specific and useful answer.



## 2. Data Collection and Preparation
Once the goal is set, relevant data is gathered from various sources. This is often the most time-consuming part of the process.

* **Collection:** Data might come from customer relationship management (CRM) systems, website analytics, sales figures, social media, or public datasets.
* **Cleaning:** Raw data is almost always messy. This step involves correcting errors, handling missing values, and removing duplicate entries.
* **Preparation (Feature Engineering):** Data is transformed into a format suitable for analysis. This might involve combining certain data points or creating new variables (called "features") that could influence the outcome. For example, instead of just using a customer's birthdate, you might create a feature for their age group.

## 3. Building the Predictive Model
This is where machine learning and statistics come into play. A predictive model is an algorithm that is "trained" on the historical data to find patterns.

The data is typically split into two parts: a **training set** and a **testing set**.
* The **training set** (usually the larger portion) is fed into the algorithm to teach it the patterns.
* The **testing set** is used to evaluate how well the model learned. By using data the model has never seen before, you can check if its predictions are accurate.

Common types of models include:
* **Regression Models:** Used for predicting continuous numerical values, like forecasting sales revenue or a house price.
* **Classification Models:** Used for predicting a category or a yes/no outcome. For instance, classifying an email as "spam" or "not spam," or determining if a customer will "churn" (leave) or "not churn."
* **Clustering Models:** Used for grouping data points into natural clusters based on their similarities, often used in customer segmentation.




## 4. Deployment and Monitoring
After a model is trained and tested, it's deployed into a live environment to make real-world predictions. For example, it could be integrated into a company's website to recommend products to users in real-time.

The work doesn't stop there. ü§öüèæ 

Models must be continuously **monitored** to ensure they remain accurate. Over time, new data might reveal new patterns, and the model may need to be retrained or updated to maintain its predictive power.

```python
import yfinance as yf
import pandas as pd

# Use distinct names for ticker strings
tcs_ticker = "TCS.NS"
infy_ticker = "INFY.NS"
nifty_ticker = "^NSEI"

start_date = "2018-01-01"
end_date = "2025-09-25"

# --- Download data into new DataFrame variables ---
tcs = yf.download(tcs_ticker, start=start_date, end=end_date, progress=False)
infy = yf.download(infy_ticker, start=start_date, end=end_date, progress=False)
nifty = yf.download(nifty_ticker, start=start_date, end=end_date, progress=False)

# --- Format dates (this part was correct) ---
start_date_formatted = pd.to_datetime(start_date).strftime('%d-%B-%Y').lower()
end_date_formatted = pd.to_datetime(end_date).strftime('%d-%B-%Y').lower()

# --- Create headings using the original ticker strings ---
tcs_heading = f"{tcs_ticker.split('.')[0]} ({start_date_formatted} to {end_date_formatted})"
infy_heading = f"{infy_ticker.split('.')[0]} ({start_date_formatted} to {end_date_formatted})"
# A better name for ^NSEI is "Nifty 50"
nifty_heading = f"Nifty 50 ({start_date_formatted} to {end_date_formatted})"


```

```python
# Interactive 2D Plotly time-series with range slider & buttons
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook_connected"  # change if needed

# prepare df safely
df = india_train.copy()
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)
if 'Date' not in df.columns:
    df = df.reset_index().rename(columns={df.columns[0]:'Date'})
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# pick columns that exist
wanted = ["TCS","INFY","NIFTY"]
available = [c for c in wanted if c in df.columns]
if not available:
    # try fuzzy matches if exact names not present
    cols = [c for c in df.columns if any(w.lower() in str(c).lower() for w in wanted)]
    available = cols[:3]

df_long = df.melt(id_vars='Date', value_vars=available, 
                  var_name='Company', value_name='Closing Price').dropna(subset=['Closing Price'])

fig = px.line(df_long, x='Date', y='Closing Price', color='Company',
              title='TCS, Infosys, Nifty 50')

# enable range slider and selectors + increase chart size
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=3, label="3m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(count=1, label="YTD", step="year", stepmode="todate"),
                dict(count=1, label="1y", step="year", stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(visible=True),
        type="date"
    ),
    hovermode='x unified',
    height=600,   # üëà increase chart height
    width=900,   # üëà optional: increase chart width
    margin=dict(l=40, r=40, t=60, b=40)  # adjust margins
)

fig.show()
```

```python
            import numpy as np
            import pandas as pd
            import matplotlib.pyplot as plt
            from sklearn.linear_model import LinearRegression
            from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error
            import yfinance as yf
            
            def simple_forecast(train_df, col_name, start_pred="2025-01-01", end_pred="2025-09-25"):
                df = train_df[["Date", col_name]].dropna().copy()
                df = df.sort_values("Date").reset_index(drop=True)
            
                # Features: trend + seasonality
                df["t"] = np.arange(len(df))
                df["doy"] = df["Date"].dt.dayofyear
                df["sin_doy"] = np.sin(2*np.pi*df["doy"]/365.25)
                df["cos_doy"] = np.cos(2*np.pi*df["doy"]/365.25)
            
                X = df[["t","sin_doy","cos_doy"]].values
                y = df[col_name].values
                model = LinearRegression().fit(X, y)
            
                # Forecast dates
                future_dates = pd.bdate_range(start=start_pred, end=end_pred)
                t_future = np.arange(len(df), len(df)+len(future_dates))
                future = pd.DataFrame({"Date": future_dates})
                future["t"] = t_future
                future["doy"] = future["Date"].dt.dayofyear
                future["sin_doy"] = np.sin(2*np.pi*future["doy"]/365.25)
                future["cos_doy"] = np.cos(2*np.pi*future["doy"]/365.25)
                future["Predicted"] = model.predict(future[["t","sin_doy","cos_doy"]].values)
            
                return future
            
            # --- Forecast NVDA, AMD, TCS, INFY ---
            
            forecast_tcs  = simple_forecast(india_train, "TCS")
            forecast_infy = simple_forecast(india_train, "INFY")
            
            # --- Download actuals for Jul-Sep 2025 ---
            
            # --- Use already cleaned data (tcs, infy) as actuals ---
            # Restrict to the period Jan‚ÄìSep 2025
            
            tcs_actual  = tcs[(tcs["Date"] >= "2025-01-01") & (tcs["Date"] <= "2025-09-25")][["Date","TCS"]].reset_index(drop=True)
            infy_actual = infy[(infy["Date"] >= "2025-01-01") & (infy["Date"] <= "2025-09-25")][["Date","INFY"]].reset_index(drop=True)
            
            
            
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import statsmodels.api as sm

# --- 1) Download data (2022‚Äì2024 for training) ---
tickers = {
    "TCS": "TCS.NS",
    "INFY": "INFY.NS",
    "NIFTY": "^NSEI",
    "USDINR": "USDINR=X",
    "OIL": "CL=F",        # Crude Oil Futures
    "SP500": "^GSPC"
}

data = {}
for name, t in tickers.items():
    df = yf.download(t, start="2022-01-01", end="2024-12-31", progress=False)
    data[name] = df[["Close"]].rename(columns={"Close": name})

# Merge into single dataset (prices)
df = pd.concat(data.values(), axis=1).dropna().reset_index()

# --- 2) Compute daily returns ---
returns = df.copy()
for col in tickers.keys():
    returns[col] = returns[col].pct_change()
returns = returns.dropna().reset_index(drop=True)

# --- Helper: regression on returns (with p-value) ---
def run_return_regression(x_col, y_col):
    X = returns[[x_col]]
    y = returns[y_col]

    # add constant for intercept
    X_sm = sm.add_constant(X)
    model = sm.OLS(y, X_sm).fit()

    beta = float(model.params[1])
    intercept = float(model.params[0])
    r2 = float(model.rsquared)
    pval = float(model.pvalues[1])

    # Scatterplot
    plt.figure(figsize=(7,5))
    plt.scatter(X, y, alpha=0.4, label="Daily returns")
    plt.plot(X, model.predict(X_sm), color="red",
             label=f"Fit line (Œ≤={beta:.3f}, R¬≤={r2:.3f}, p={pval:.3g})")
    plt.xlabel(x_col); plt.ylabel(y_col)
    plt.title(f"{y_col} vs {x_col} (returns)")
    plt.legend(); plt.grid(True); plt.show()

    # Nicely formatted text output
    print(f"Regression Results: {y_col} ~ {x_col}")
    print(f"  Beta (slope): {beta:.6f}")
    print(f"  Intercept:    {intercept:.6f}")
    print(f"  R¬≤:           {r2:.6f}")
    print(f"  p-value:      {pval:.6g}")

    return beta, intercept, r2, pval
```

```python


print("\nNIFTY vs SP500")
print(run_return_regression("SP500", "NIFTY"))
```
